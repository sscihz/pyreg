```python
#matplotlib版本
#本来是画图原教旨主义，但是感觉这些功能还是seaborn实现起来方便一些。。。
#为了方便我在下面重新用seaborn写一个
def paneltreatment(data,treatment_var_li,x_var,y_var,mode="treatment_bi"):
    ## treatment matrix list
    tmli = []
    for tm in treatment_var_li:
        treatment_ma = data.pivot_table(columns=x_var, index=y_var, values=tm)
        if mode == "missing":
            treatment_ma[~treatment_ma.isnull()] = 1  # not nan
            treatment_ma[treatment_ma.isnull()] = 0   # nan
        tmli.append(treatment_ma)
    
    ##color map
    ## for categorical, max is five
    category_colormap = ['#CD5C5C','#F08080','#FA8072','#E9967A','#FFA07A']
    bi_colormap = ['#3D71A0', '#B70050']
    
    ## set x and y
    x = len(set(data[x_var]))
    y = len(set(data[y_var]))
    xlb = sorted(list(set(data[x_var])))
    ylb = sorted(list(set(data[y_var])))
    
    ## draw plots 
    num = len(tmli)//4 + 1
    fig, axes = plt.subplots(ncols=4, nrows=num, constrained_layout=True,figsize=(10,10),dpi=400)
    cmap = matplotlib.colors.ListedColormap(bi_colormap)

    for tm in range(len(tmli)):
        treatma = tmli[tm]
        ax = axes[tm]
        ax.set_facecolor('white')
        ax.imshow(treatma, cmap=cmap, origin="lower", extent=(0,x,0,y))
        ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=.1)
        ## set labels 
        ax.set_xticks(np.arange(0, x, 1))
        ax.set_yticks(np.arange(0, y, 1))
        ax.set_yticklabels(ylb,rotation=0, fontsize=2)
        ax.set_xticklabels(xlb,rotation=90, fontsize=2)


    plt.show()
```

```python
paneltreatment(capacity,['demo'],'year','country','treatment_bi')
```
